```python
# data processing for regression
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.get_dummies(df,drop_first = True)
```

```python
df
```

```python
df['sex_male'].replace({True:1,False:0},inplace = True)
```

```python
df['region_northwest'].replace({True:1,False:0},inplace = True)
df['region_southeast'].replace({True:1,False:0},inplace = True)
df['region_southwest'].replace({True:1,False:0},inplace = True)
```

```python
df
```

```python
df.describe()
```

```python
sns.pairplot(df,diag_kind = 'kde')
```

```python
# model building
```

```python
import statsmodels.formula.api as sm
```

```python
lm = sm.ols(formula = 'charges~ age+bmi+children+sex_male+smoker_yes+region_northwest+region_southeast+region_southwest',data = df).fit()
lm.params
```

```python
df.info()
```

```python
print(lm.summary())
```

```python
lm2 = sm.ols(formula = 'charges~ age+bmi+children+sex_male+smoker_yes+region_southeast+region_southwest',data = df).fit()
lm2.params
```

```python
print(lm2.summary())
```

```python
lm3 = sm.ols(formula = 'charges~ age+bmi+children+sex_male+smoker_yes+region_southeast',data = df).fit()
lm3.params
```

```python
print(lm3.summary())
```

```python
# combining the effect of smoking habit and region by creating theree more dummy-variables as below:

df['smoker_northwest'] = df['smoker_yes']*df['region_northwest']
df['smoker_southeast'] = df['smoker_yes']*df['region_southeast']
df['smoker_southwest'] = df['smoker_yes']*df['region_southwest']
```

```python
df
```

```python
lm4 = sm.ols(formula = 'charges~ age+bmi+sex_male+smoker_yes+region_northwest+region_southeast+region_southwest+smoker_northwest+smoker_southeast',data = df).fit()
lm4.params
```

```python
print(lm4.summary())
```

```python
lm5 = sm.ols(lm4 = sm.ols(formula = 'charges~ age+bmi+sex_male+smoker_yes+region_southeast+region_southwest+smoker_southeast',data=df).fit())
lm5.params
```

```python
from sklearn.model_selection import train_test_split
```

```python
data_train,data_test = train_test_split(df,test_size=0.10,random_state = 22,shuffle = True)
```

```python
data_train.columns
```

```python
data_train
```

```python
data_test
```

```python
lm1 = sm.ols(formula = 'charges ~age+bmi+children+sex_male+smoker_yes+region_northwest+region_southeast+region_southwest+smoker_northwest+smoker_southeast',data = data_train).fit()
lm1.params
```

```python
print(lm1.summary())
```

```python
lm2 = sm.ols(formula = 'charges ~age+bmi+children+sex_male+smoker_yes+region_southeast',data = data_train).fit()
lm2.params
```

```python
print(lm2.summary())
```

```python
ypred_train = lm2.predict(data_train)
ypred_train
```

```python
ypred_test = lm2.predict(data_test)
ypred_test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(data_train['charges'],ypred_train,squared = False)
```

```python
data_train['charges'].mean()
```

```python
cient_of_variation_train = mean_squared_error(data_train['charges'],ypred_train,squared= False)/data_train['charges'].mean()
cient_of_variation_train
```
